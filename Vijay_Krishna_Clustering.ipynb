# Import required packages and libraries

```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
Trans_df = pd.read_csv('C://Users/krivijay/OneDrive - Hewlett Packard Enterprise/Desktop/Vijay/Personal/Zeotap-Intermship-Data Science/Dataset/Transactions.csv')
Trans_df.head()
```

```python
#Scatter Plot
plt.scatter(Trans_df['TotalValue'],Trans_df['Price'])
```

#K-Means

```python
km = KMeans(n_clusters=3)
km
```

```python
#KM-fit TotalValue and Price
y_pred = km.fit_predict(Trans_df[['TotalValue','Price']])
y_pred
```

```python
#Check th clustering reslut
Trans_df['cluster'] = y_pred
Trans_df.head()
```

```python
# return all the rows where respective clusters 
df1 = Trans_df[Trans_df.cluster==0]
df2 = Trans_df[Trans_df.cluster==1]
df3 = Trans_df[Trans_df.cluster==2]

plt.scatter(df1.TotalValue,df1.Price).color='green'
plt.scatter(df2.TotalValue,df2.Price).color='red'
plt.scatter(df3.TotalValue,df3.Price).color='black'

plt.xlabel('Price')
plt.ylabel('TotalValue')
plt.show
```
